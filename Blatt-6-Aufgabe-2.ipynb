Aufgabe 2: Praxisaufgabe: Exponentialverteilung (2 Punkte + 2 Bonuspunkte)

Eine exponentialverteilte Zufallsvariable X nimmt Werte größer oder gleich Null an. Sie wird beschrie-
ben durch die Dichtefunktion f (x) = e−x, x ≥ 0. Die Verteilungsfunktion lautet F (x) = 1 − e−x,
x ≥ 0.



a) (2 Punkte) In der Python-Bibliothek random gibt es eine Methode um eine exponentialverteilte
Zufallsvariable X zu erzeugen. Setzen Sie den Parameter lambd=1. Erstellen Sie ein Diagramm,
das ein Histogramm von 10000 Realisierungen der exponentialverteilten Zufallsvariablen X
zeigt. Verwenden Sie für das Histogramm mindestens 10 sinnvoll gewählte Klassen. Stellen Sie
im selben Diagramm die zugehörige Dichtefunktion dar. Wählen Sie für Ihr Diagramm einen
sinnvollen Definitionsbereich für die Dichtefunktion. Berechnen Sie das arithmetische Mittel
der Daten.


```python
import numpy as np
import matplotlib.pyplot as plt
import random

lambd = 1
generierte_daten = [random.expovariate(lambd) for _ in range(10000)]

mean_wert = np.mean(generierte_daten)

plt.hist(generierte_daten, bins=10, density=True, alpha=0.6, color='g')

#Erstellung der Dichtefunktion
x = np.linspace(0, max(generierte_daten), 1000)
y = np.exp(-x)
plt.plot(x, y, 'r', linewidth=2)

plt.title("Histogramm und Dichtefunktion der Exponentialverteilung")
plt.xlabel("Werte von X")
plt.ylabel("Dichte")
plt.show()

mean_wert
```

b) (2 Bonuspunkte) Verwenden Sie die Inversionsmethode zur Simulation der exponentialverteil-
ten Zufallsvariablen X. Erläutern Sie schriftlich in einer Markdown-Zelle Ihre Vorgehensweise.
Stellen Sie 10000 Realisierungen dieser Zufallsvariablen in einem Diagramm als Histogramm
zusammen mit der zugehörigen Dichtefunktion dar. Wählen Sie für Ihr Diagramm einen sinn-
vollen Definitionsbereich für die Dichtefunktion. Berechnen Sie das arithmetische Mittel der
Daten.

```python
generierte_daten_im_Intervall = np.random.uniform(0, 1, 10000)

# Anwendung der Inversionsmethode zur Simulation der exponentialverteilten Zufallsvariable X
# X = -ln(U) mit U als gleichverteilte Zufallszahl
simulierte_daten = -np.log(generierte_daten_im_Intervall)

mean_simulated_data = np.mean(simulierte_daten)

plt.hist(simulierte_daten, bins=10, density=True, alpha=0.6, color='b')
plt.plot(x, y, 'r', linewidth=2)

plt.title("Histogramm und Dichtefunktion (Inversionsmethode)")
plt.xlabel("Werte von X")
plt.ylabel("Dichte")
plt.show()

mean_simulated_data
```

Vorgehensweise

Ziel ist es, 10000 Realisierungen einer exponentialverteilten Zufallsvariablen \( X \) mit der Inversionsmethode zu simulieren und diese in einem Diagramm als Histogramm darzustellen. Zusätzlich soll die zugehörige Dichtefunktion im Diagramm abgebildet und das arithmetische Mittel der simulierten Daten berechnet werden. Wir generieren 10000 gleichverteilte Zufallszahlen im Intervall \([0, 1]\) und Jede dieser Zahlen wird dann mit der Inversionsmethode transformiert, um die entsprechenden Realisierungen von \( X \) zu erhalten. Die simulierten Daten werden in einem Histogramm mit 10 Klassen dargestellt. Diese Anzahl der Klassen bietet einen guten Kompromiss zwischen Detailgenauigkeit und Übersichtlichkeit. Zusätzlich wird die Dichtefunktion der Exponentialverteilung im selben Diagramm dargestellt, um einen visuellen Vergleich zwischen der simulierten und der theoretischen Verteilung zu ermöglichen.

Das Ergebnis ist ein Histogramm, das die Verteilung der mit der Inversionsmethode simulierten Daten zeigt, zusammen mit der Dichtefunktion der Exponentialverteilung und dem berechneten arithmetischen Mittel, das die Genauigkeit der Simulation bestätigt.
