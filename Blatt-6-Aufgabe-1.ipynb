Aufgabe 1: Praxisaufgabe: Simulation der Augensumme durch die Inversionsmethode (2 Punkte)

In der Python-Bibliothek random gibt es die Methode random, um eine gleichverteilte Zufallsvariable
zu erzeugen mit 0 ≤ X < 1. Nutzen Sie X, um die Wahrscheinlichkeiten für die möglichen Augen-
summen (2 bis 12) von 2 geworfenen Würfeln zu simulieren. Erzeugen Sie 10000 Realisierungen von
X und stellen die relativen Häufigkeiten für die simulierten Augensummen in einem Balkendiagramm
dar. Stellen Sie in einem zweiten Balkendiagramm den Fehler der simulierten Wahrscheinlichkeiten
dar, d.h. die Differenz der relativen Häufigkeiten zu den exakten Wahrscheinlichkeiten

```python
import random
import math
import matplotlib.pyplot as plt

anzahl_der_simulationen = 10000

ergebnis = {i: 0 for i in range(2, 13)}

for _ in range(anzahl_der_simulationen):
  simuliere_2_wuerfe =  math.trunc( random.random() * 11) + 2
  ergebnis[simuliere_2_wuerfe] += 1

keys = list(ergebnis.keys())
relative_haeufigkeit = [freq / anzahl_der_simulationen for freq in ergebnis.values()]

exakte_berechnung = 1 / len(keys)
exakte_wahrscheinlichkeit = {}

for i in keys:
  exakte_wahrscheinlichkeit[i] = exakte_berechnung

differenz_zur_relativen_haeufigkeit = []

for i in relative_haeufigkeit:
  differenz_zur_relativen_haeufigkeit.append(abs(i - exakte_berechnung))


plt.figure(figsize=(10,6))
plt.bar(keys, relative_haeufigkeit, color="blue")
plt.xlabel("Würfel Ergebnis")
plt.ylabel("Realtive Häufigkeit")
plt.xticks(keys)



plt.figure(figsize=(10,6))
plt.bar(keys, differenz_zur_relativen_haeufigkeit, color="red")
plt.xlabel("Würfel Ergebnis")
plt.ylabel("Differenz zur exakten Wahrscheinlichkeit")
plt.xticks(keys)

```
