Aufgabe 1: Praxisaufgabe: Numerische Integration (3 Punkte)

Die Datei normal.txt enthält 10000 Realisierungen einer normalverteilten Zufallsvariablen X ∼
N (μ, σ2). Schätzen Sie die Parameter μ und σ2 aus den Daten und geben Sie die Dichtefunktion
zusammen mit dem Histogramm der Daten graphisch aus. Berechnen Sie die Wahrscheinlichkeit
P (X > 5) durch Verwendung der Funktion norm.cdf. Vergleichen Sie den Wert des Modells mit dem
empirischen Wert, d.h. bestimmen Sie den relativen Anteil der Daten, der größer als 5 ist.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

daten = np.loadtxt('normal.txt')

mu_schaetzung = np.mean(daten)
sigma_schaetzung = np.var(daten)

plt.hist(daten, bins=30, density=True, alpha=0.6, color='g')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_schaetzung, np.sqrt(sigma_schaetzung))
plt.plot(x, p, 'k', linewidth=2)
title = "Ergebnis: mu = {:.2f},  sigma^2 = {:.2f}".format(mu_schaetzung, sigma_schaetzung)
plt.title(title)

plt.show()

p_x_größer_5 = 1 - norm.cdf(5, mu_schaetzung, np.sqrt(sigma_schaetzung))
empirischer_wert = np.mean(daten > 5)

output = f"Schätzung des Mittelwerts (mu): {mu_schaetzung}\n" \
         f"Schätzung der Varianz (sigma^2): {sigma_schaetzung}\n" \
         f"Wahrscheinlichkeit P(X > 5): {p_x_größer_5}\n" \
         f"Empirischer Wert (Daten > 5): {empirischer_wert}"
print(output)
```
